```python
le = LabelEncoder()
categorical_feature_mask = hp.dtypes==object#select the columns with categorical dtype
categorical_cols = hp.columns[categorical_feature_mask].tolist()
hp[categorical_cols] = hp[categorical_cols].apply(lambda col: le.fit_transform(col.astype(str)))
hp.head(10)
```

```python
hp['LotFrontage'].fillna(np.mean(hp['LotFrontage']),inplace=True)
```
